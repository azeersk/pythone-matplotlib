```python
#matplotlib
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
x = np.arange(1,1)
y = x*2
a = plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("line plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid()
plt.show()
```

```python
x = np.arange(1,7)
y = x*2
plt.plot(x,y,color = 'green',linestyle = '-.',linewidth = 2)
plt.show()
```

```python
x = np.arange(1,9)
y1 = x*2
y2 = x*3
plt.plot(x,y1,color = 'r',linestyle = ":",linewidth = 2)
plt.plot(x,y2,color = 'y',linestyle = "-.", linewidth = 2)
plt.grid(True)
plt.show()
```

```python
#adding sub-plots
x = np.arange(1,9)
y1 = x*2
plt.subplot(1,2,1)
plt.plot(x,y1,color = 'r',linestyle = "-.",linewidth = 3)
y2 = x*3 
plt.subplot(1,2,2)
plt.plot(x,y2,color = 'g',linestyle = "--",linewidth = 5)
plt.show()
```

```python
#bar plot 
student = {"bob":87,"matt":56,"sam":27}
names = list(student.keys())
values = list(student.values())
plt.bar(names,values)
plt.grid()
plt.show()
```

```python
student = {"bob":40,"matt":56,"sam":27}
names = list(student.keys())
values = list(student.values())
plt.title("bar groph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.barh(names,values,color = 'r')
plt.show()
```

```python
#scatter plot
x = [1,2,3,4,5,6,3,2,1]
y1 = [10,20,30,40,50,60,70,80,90]
y2 = [10,20,30,40,50,60,70,80,90]
plt.scatter(x,y1,marker = ",",c = 'g',s = 500)
plt.scatter(x,y2,marker = "*",c = 'r',s = 100)
plt.show()
```

```python
#histagram
data = [1,1,1,1,2,3,3,3,3,2,2,2,4,4,5,6,6,6,7,7,7,8,8,8,8,9,9,9,9]
plt.hist(data,color = 'r',bins = 5)

plt.show()
```

```python
azeer = pd.DataFrame({'markes':[90,80,70,76,86,86,54]})
plt.hist(azeer,color = 'g',bins = 5)
plt.show()
```

```python
#box plot
one = [1,2,3,4,5,6,7,6,5,4,3,2,1,9,3,4,5]
two = [1,2,3,4]
three = [1,4,5,2,3,6,9,4,3,2,5]
ar = list([one,two,three])
plt.boxplot(ar)
plt.show()
```

```python
#violin plot
one = [1,2,3,4,5,6,7,6,5,4,3,2,1,9,3,4,5]

three = [1,4,5,2,3,6,9,0,0,0,0,4,3,2,5]
two = [1,2,3,4,9,5,6,7,8]
ar = list([one,three,two])
plt.violinplot(ar)
plt.show()
```

```python
#pie chart
fruit = ['apple','orange','banana','mango','guava']
quantity = [90,80,60,99,50]
plt.pie(quantity,labels = fruit, autopct = "%0.1f%%",colors = ['red','orange','yellow','green','blue'],shadow = True)
plt.show()
```

```python
#doughnut chat
fruit = ['apple','orange','mango']
quantity = [35,35,35]
plt.pie(quantity,labels = fruit,radius = 1.1,colors = ['red','yellow','green'])
plt.pie([1],colors = ['blue'],radius = 0.4)
plt.show()
```
